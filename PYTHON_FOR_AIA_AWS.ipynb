```python
# ---> Important things in functions !

# 1) ----> Anonymous Functions (Lambda Functions):
            #Lambda functions are small, anonymous functions defined using the 'lambda' keyword.
add = lambda a,b:a+b
re = add(3,4) # output is 7
# Using this Lambda function we can combine this with the 'map' or 'filter' or etc where we can save the time and space and the execution time !

# 2) ----> Higher-Order Functions:
            #Python treats functions as first-class citizens , allowing functions to be passed as arguments to other functions or returned from functions.
def apply_operation(func, a, b):
    return func(a, b)

result = apply_operation(add, 2, 3)  # Output: 5 #In the place of the 'add' we can pass a actual function name then that function will call those respective function 
#Internally.

# 3) ----> Recursive Functions:
            #Functions can call themselves, allowing for recursive algorithms.
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

result = factorial(5)  # Output: 120

# 4) ----> Functions as Objects:
            #Functions are first-class objects in Python.
            #They can be assigned to variable, stored in data structures, passed as arguments, returned from
def greet():
    return "Hello, world!"

def bye():
    return "Goodbye, world!"

my_functions = [greet, bye]
print(my_functions[0]()) # Output: Hello, world!
print(my_functions[1]()) # Output: Goodbye, world!
#----------------------------------------------------------------------------------
def greet(name):
    return f"Hello, {name}!"

# Functions can be assigned to variables
greet_someone = greet

# Functions can be stored in data structures
function_list = [greet, greet_someone]

# Functions can be passed as arguments to other functions
def call_function(func, arg):
    return func(arg)

print(call_function(greet, "Alice")) # Output: Hello, Alice!

# Functions can be returned as values from other functions
def create_greet_function(greeting):
    def greet(name):
        return f"{greeting}, {name}!"
    return greet

greet_with_hi = create_greet_function("Hi")
print(greet_with_hi("Bob")) # Output: Hi, Bob!

# ----> Nested Functions:
        #Functions can be defined inside other functions, enabling encapsulation and modular design.
def outer_function(x):
    def inner_function(y):
        return x + y
    return inner_function

closure = outer_function(3)
result = closure(5)  # Output: 8
```

## ----> ℃lass ℳethods:


            Class methods are a type of method in Python that are bound to the class rather than its instances (objects).

## ----> ₰tatic ℳethods

        
        In Python, a static method is a method that belongs to a class but does not access or modify the class or instance state. 
        Unlike regular methods in a class, static methods do not take the implicit first argument self (instance reference) or 
        cls (class reference).
        
        To define a static method in Python, use the @staticmethod decorator before the 

```python
#Working with static methods 

#NOTE:--
"""     In a static method, by default, you cannot access instance variables (attributes bound to a specific instance of the class using self). 
        Static methods are designed to be independent of instance state.

        However, if you have a reference to an instance within a static method, you can access its instance variables by using that reference. 
        But this is not a recommended practice as it breaks the concept of a static method, 
        which is meant to be independent of instance-specific data. """

#Example problem --
class MyClass:
    def __init__(self, value):
        self.value = value

    @staticmethod
    def my_static_method(obj):
        # Accessing instance variable via object reference
        print(obj.value)

# Create an instance of MyClass
obj = MyClass(42)

# Call the static method and pass the instance as an argument
MyClass.my_static_method(obj)  # Output: 42

```

```python

#Example of '@static method'
from functools import reduce
class items:
    all=[]
    def __init__(self,name:str,price:float,units:int):
        self.name = name
        self.price = price
        self.units = units
        items.all.append(self)

    @staticmethod 
    def stock_value(k1):
        op =[]
        for i in k1:
            d = reduce(lambda x,y: x*y,i)
            op.append(d)
        return sum(op)
        
        

o1 = items('samsung',20300,12)
o2 = items('apple',40223,5)
tot=[]
for i in items.all:
    i1 = (i.price)
    i2 = (i.units)
    tot.append([i1,i2])

res = items.stock_value(tot)
print(f"The over all worth of this mobile company is {res:,} Rs")

```

## ----> Ǐnheritance

                Inheritance in Python is a mechanism that allows a class (referred to as a child or subclass) to inherit properties, 
                methods, and attributes from another class (referred to as a parent or superclass). It enables the creation 
                of a new class with the characteristics of an existing class, promoting reusability and establishing a 
                hierarchical relationship between classes.

```python
#This is the best example !!!
class items:
    all=[]
    def __init__(self,name:str,price:float,units:int):
        self.name = name
        self.price = price
        self.units = units
        items.all.append(self)
    

        
class Phone(items):#In this class  i am trying to add a new attribute 'broken_phones' so i have used super(). to instantiate all the attributes and methods from items class
    def __init__(self,name:str,price:float,units:int,broken_phones:int):
        super().__init__(name,price,units)
        self.broken_phones = broken_phones#we need to add only the new one 
    def __repr__(self):
        return f"[{self.name},{self.price},{self.units}]"


    @classmethod
    def clas(cla):
        for i in cla.all:
            if i:
                i.units -= i.broken_phones
        


o1 = Phone('samsung',20300,12,3)
o2 = Phone('apple',40223,5,2)
Phone.clas()
print(Phone.all)

```

```python
#Another example for inheritance
# Parent class
class Animal:
    def __init__(self, species):
        self.species = species

    def sound(self):
        pass  # To be implemented in child classes

# Child classes inheriting from Animal
class Dog(Animal):
    def sound(self):
        return "Woof!"

class Cat(Animal):
    def sound(self):
        return "Meow!"

# Creating instances of child classes
dog = Dog("Canine")
cat = Cat("Feline")

# Accessing attributes and methods inherited from the parent class
print(dog.species)  # Output: Canine
print(cat.species)  # Output: Feline

print(dog.sound())  # Output: Woof!
print(cat.sound())  # Output: Meow!


# Calling the parent class classmethod from the child class!
class Parent:
    @classmethod
    def class_method(cls):
        print("This is a class method from the Parent class")

class Child(Parent):
    def call_parent_class_method(self):
        super().class_method()  # Calling parent class's class_method using super()

child_obj = Child()
child_obj.call_parent_class_method()
```

## ----> Encapsulation {Getter,Setter}

        @property is a built-in decorator in Python used to define properties in a Python class. 
        It allows you to define methods that can be accessed like attributes, providing a 
        way to encapsulate instance attributes and controlling their access.

        When you use @property decorator on a method within a class, you can access the 
        method as an attribute without using parentheses. It provides a way to define 
        getter, setter, and deleter methods for a specific attribute, 
        enabling you to control its access, modification, and deletion.

```python
#Implementation of the Encapsulation !!! ----> best example about Getter and Setter !!!
class mot:
    def __init__(self,name:str,age:float,sex:str):
        #This assert statements are used to validate the data at the object creation to the class
        assert age >=0,f"The age that you have entered should not be less than ZERO {age}"
        assert len(sex) == 1 and sex.upper() in "MF","The sex data must not be 'M' or 'F' and the length must be greater than 0!"
        
        self.__name = name # It is a private variable !
        self.__age = age # It is a private variable !
        self.__sex = sex # It is a private variable !

        

    #This getters and setters will act as an access specifiers an as well as a data validations also !
    @property
    def name(self):#This is the Getter for name!
        return self.__name

    @name.setter
    def name(self,value):#This is the Setter !
        self.__name = value



    @property
    def age(self):#This is the Getter for name!
        return self.__age

    @age.setter
    def age(self,value):#This is the Setter !
        if value >=0:
            self.__age = value
        else:
            raise ValueError("The age must be greater then the ZERO")



    @property
    def sex(self):#This is the Getter for name!
        return self.__sex

    @sex.setter
    def sex(self,value):#This is the Setter !
        if len(value) == 1:
            if value in "MmFf":
                self.__sex = value
            else:
                raise ValueError("The sex data must be in this \"MmFf\" character !")
        else:
            raise ValueError("The age must be greater then the ZERO")



m1 = mot('vinay',33,'M')
print(m1.name) #AttributeError: 'mot' object has no attribute 'name' because  we haven't  created the setter and getter's !
m1.name = 'sam' #using the setter we can now change the data in that !
print(m1.name)
print(m1.age) #From the object 
m1.age = 78
print(m1.age) #Updated by the use of the setter !
print(m1.sex)
m1.sex="m"
print(m1.sex) #Updated by the use of the setter !
m2 = mot('kumar',35,'f')


```

```python
def select_archer(score_list):
    selected_archers = []
    for score in score_list:
        archer_details = score.split(',')
        archer_id = archer_details[0]
        scores = list(map(int, archer_details[1:]))
        selected_scores = []
        total_score = 0
        for s in scores:
            if total_score + s >= 50:
                break
            total_score += s
            selected_scores.append(s)
        if total_score >= 50:
            selected_archers.append({archer_id: selected_scores})
    return selected_archers

def extract_id(selected_list):
    return [list(d.keys())[0] for d in selected_list]

# main method
archers_score = []
no_of_archers = int(input("Enter the no of archers:\n"))
for _ in range(no_of_archers):
    archers_score.append(input("Enter the archer id and 10 scores of round1 as comma-separated values:\n"))
selected = select_archer(archers_score)
if not selected:
    print("No one selected")
else:
    print("Selected Archers Id's:\n")
    for id in extract_id(selected):
        print(id)

```
